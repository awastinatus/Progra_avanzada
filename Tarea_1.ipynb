*  ¿Cuál es la diferencia entre recursividad e iteración, y cómo se relaciona esto con la notación big 𝑂?

* La recursividad es una técnica en la que una función se llama a sí misma para resolver un problema dividiéndolo en subproblemas más pequeños. En cambio, la iteración se refiere a repetir un bloque de código varias veces utilizando estructuras como bucles (for, while). En términos de notación big O, la recursividad puede llevar a una complejidad temporal mayor si no se implementa cuidadosamente, debido a la sobrecarga de las llamadas de función repetidas. La iteración suele ser más eficiente en términos de uso de memoria y tiempo de ejecución, aunque ambos enfoques pueden tener la misma complejidad big O en muchos casos.

* ¿Cómo se puede determinar la complejidad temporal de un algoritmo recursivo?

* Primero se identifica el caso base y su complejidad, luego se determinar el número de llamadas recursivas en cada nivel, despues hay que calcular el trabajo realizado en cada llamada recursiva, excluyendo las llamadas recursivas en sí, se establece una relación de recurrencia que describa cómo crece la función con el tamaño de la entrada y se resuelve la relación de recurrencia para obtener una expresión cerrada de la complejidad.





```python
#Clase capaz de calcular la cantidad posible de caminos entre A y B, utilizando forma recursiva y forma de combinatoria dinamica
class CaminosPCB:
    def __init__(self, N, M):
        self.N = N  #Número de filas
        self.M = M  #Número de columnas

    def caminos_recursivo(self, i, j):
        if i == self.N - 1 and j == self.M - 1:
            return 1  #Solo un camino cuando estamos en la celda final
        if i >= self.N or j >= self.M:
            return 0  #Fuera de los límites

        #Llamadas recursivas: hacia la derecha (i, j+1) y hacia abajo (i+1, j)
        return self.caminos_recursivo(i, j + 1) + self.caminos_recursivo(i + 1, j)

    def caminos_dinamico(self):
        #Inicializamos una tabla de NxM con ceros
        dp = [[0] * self.M for _ in range(self.N)]

        #Solo hay un camino posible para la primera fila y la primera columna
        for i in range(self.N):
            dp[i][0] = 1
        for j in range(self.M):
            dp[0][j] = 1


        for i in range(1, self.N):
            for j in range(1, self.M):
                dp[i][j] = dp[i - 1][j] + dp[i][j - 1]

        #resultado está en la última celda
        return dp[self.N - 1][self.M - 1]



```
